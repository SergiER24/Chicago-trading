```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import pandas as pd
```

Constantes y calculo de Variables

```python
Q_array=np.linspace(0,0.5,10) #array caudal values
K_array=[0.6*9] #array K values
###PVC 1
constants_pvc_1={'densidad':1000,'viscosidad_dinamica':1e-3,
           'diametro':0.1524,'rugosidad':1.5e-4,'altura':75,
           'longitud':488.22,'gravedad':35316} #constats
V_array_pvc_1=[(i/((np.pi/4)*(constants_pvc_1['diametro']**2)))for i in Q_array] #loop 4 V_array
reynolds_array_pvc_1=[((constants_pvc_1['densidad']*i*constants_pvc_1['diametro'])/constants_pvc_1['viscosidad_dinamica']) for i in V_array_pvc_1] #reynolds loop
relative_roughness_pvc_1=constants_pvc_1['rugosidad']/constants_pvc_1['diametro']


friction_f_pvc_1=[0,0.020210766444459453,
                  0.019894737369517468,
                  0.019785338954984754,
                  0.019729820827443022,
                  0.01969623964882218,
                  0.01967373757999125,
                  0.019657607892884968,
                  0.01964547932575464,
                  0.019636027335629764]
H_pvc1=[]
for i in range(0,len(Q_array)):
    H_pvc1.append(constants_pvc_1['altura']+(friction_f_pvc_1[i]*constants_pvc_1['longitud']*(Q_array[i]**2)/(constants_pvc_1['diametro']*2*constants_pvc_1['gravedad']*((np.pi/4)*(constants_pvc_1['diametro']**2))**2))+K_array[0]*((Q_array[i]**2)/(constants_pvc_1['diametro']*2*constants_pvc_1['gravedad']*((np.pi/4)*(constants_pvc_1['diametro']**2))**2)))
      
#Grafica
    
plt.figure(figsize=(10,6),dpi=120,facecolor='black')
plt.style.use('dark_background')

#regresion eficiencia    
coefficients = np.polyfit(H_pvc1, Q_array, 2)
polynomial = np.poly1d(coefficients)
#Cuadrar de valores
y_fit=np.linspace(min(H_pvc1), max(H_pvc1), 100)
x_fit=polynomial(y_fit)
df = pd.read_csv('Default Dataset (2).csv', header=None)
df.columns = ["1", "2"]
df['1'] = df['1'] / 1000
plt.plot(x_fit, y_fit, label="Sistema", color="r") #regresion eficiencia
plt.plot(df["1"],df["2"],label="Bomba",color="b")
plt.scatter(0.19, 75.2, color='g', label='Intersección (0.19,75.2)',zorder=5)

plt.grid()
plt.title("Curva del sistema tubo PVC 12 pulgadas y bomba" )
plt.ylabel("Head (m)")
plt.xlabel("Flow (m3/min)")
plt.legend()
```
